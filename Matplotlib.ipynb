```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Line Plot

```python
x = ["mon", "tue", "wed", "thu", "fri", "sat", "sun"]
y= [10, 12, 15, 14, 13, 17, 20]

```

```python
plt.plot(x,y)
plt.xlabel('Day of the Week')
plt.ylabel('Sales per Day')
plt.title('Bakery Sales Over a Week')
plt.show()
```

```python
months = [1,2,3,4,5]
sales = [200, 255, 325, 480, 515]
```

```python
plt.plot(months,sales, color='blue', linestyle='--',linewidth=2, marker='o', label='2025 Sales data')
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.grid(True)
plt.legend(loc='upper left', fontsize='small')
# plt.xlim(1,5)
# plt.ylim(0,550)
plt.xticks([1,2,3,4,5], ['Jan', 'Feb', 'Mar', 'Apr', 'May'])
plt.title('Company Sales Data')
plt.show()
```

# Barchart

```python
product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]
```

```python
plt.barh(product, sales, color=['red', 'blue', 'green', 'orange'],label='Sales 2025')
plt.xlabel('Products')
plt.ylabel('Sales in USD')
plt.title('Product Sales Comparison')
plt.legend()
plt.show()
```

# Pie Chart

```python
regions = ['North', 'South', 'East', 'West']
revenue = [2500, 3000, 2000, 1500]
```

```python
plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold', 'skyblue', 'lightgreen', 'lightcoral'])
plt.title('Revenue Distribution by Region')
plt.show()
```

# Histogram

```python

scores = [55, 67, 45, 70, 89, 90, 76, 88, 92, 60, 73, 85, 95, 78, 82]
```

```python
plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')
plt.title('Distribution of Student Scores')
plt.show()
```

# Scatter Plot
understanding the corelation between two variables.

```python

hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
test_scores = [50, 55, 65, 70, 75, 80, 85, 90]
```

```python
plt.scatter(hours_studied, test_scores, color='green', marker='D', label='Student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.title('Hours Studied vs Test Scores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3,4,5], [2,3,4,5,6], color='red', marker='o', label='Class A')
plt.scatter([1,2,3,4,5], [3,4,2,5,7], color='blue', marker='s', label='Class B')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.title('Comparison of Two Classes')
plt.legend()
plt.grid(True)
plt.show()
```

# SubPlots
1. For comparing multiple charts
2. Easy to analyse related datasets together
3. used in EDA a lot

```python
x = [1,2,3,4]
y = [10,20,15,25]
```

```python
plt.subplot(2,2,1)  # (1-rows, 2-columns, 1-subplot)
plt.plot(x,y, color='blue', marker='o')
plt.title('Line Chart')

plt.subplot(2,2,2)  # (1-rows, 2-columns, 2-subplot)
plt.bar(x,y, color='orange')
plt.title('Bar Chart')

plt.subplot(2,2,3)  # (2-rows, 1-columns, 1-subplot)
plt.pie([25,35,25,15], labels=['A','B','C','D'], autopct='%1.1f%%')
plt.title('Pie Chart')

plt.subplot(2,2,4)  # (2-rows, 1-columns, 2-subplot)
plt.hist([5,15,25,35,45,55,65,75,85,95], bins=5, color='green', edgecolor='black')
plt.title('Histogram')



plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot([1, 2, 3, 4], [10, 20, 15, 25], color='green', marker='o')
axs[0, 0].set_title('Line Chart')

axs[0, 1].bar([1, 2, 3, 4], [15, 25, 10, 30], color='purple')
axs[0, 1].set_title('Bar Chart')

axs[1, 0].pie([20, 30, 25, 25], labels=['W', 'X', 'Y', 'Z'], autopct='%1.1f%%')
axs[1, 0].set_title('Pie Chart')

axs[1, 1].hist([10, 20, 30, 40, 50, 60, 70, 80, 90], bins=4,  edgecolor='black')
axs[1, 1].set_title('Histogram')

fig.suptitle('Various Charts in Subplots', fontsize=16)

plt.tight_layout()
plt.show()
```
