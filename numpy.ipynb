```python
# IN PYTHON, WE CAN EASILY JOIN PANDAS DATAFRAME USING THE JOIN() METHOD. THIS WILL JOIN THE COLUMNS OF THE TWO DIFFERENT DATAFRAMES.
```

```python
# LET US SEE AN EXAMPLE AND CREATE TWO DATAFRAME AND JOIN THEM.
```

```python
import pandas as pd

#DATASET

data1 = {
  'Id': ["S01","S02","S03","S04","S05"],
  'Student': ["amit", "john", "jocob", "david", "steve"],
  'Roll': [101, 102, 103, 104, 105]
}

data2 = {
  'Rank' : [1, 4, 3, 5, 2],
  'Marks': [95, 70, 80, 60, 90]
}

#DATASET

dataFrame1 = pd.DataFrame(data1)
print("DataFrame1 = \n", dataFrame1)
dataFrame2 = pd.DataFrame(data2)
print("/nDataFrame2 = \n", dataFrame2)

# JOIN TWO DATAFRAMES
resDF = dataFrame1.join(dataFrame2)
print("\njoining two DataFrames\n", resDF)
```

```python
import pandas as pd

#DATASET
data1 = {
    'Id': ["S01","S02","S03","S04","S05"],
    'Student': ["amit", "john", "jocob", "david", "steve"],
    'Roll': [101, 102, 103, 104, 105]
}

data2 = {
  'Id': ["S06","S07","S08"],
    'Student': ["Ben", "kane", "Rohit"],
    'Roll': [106, 107, 108]
}

#Dataframe

dataFrame1 = pd.DataFrame(data1, index=['Student1','Student2','Student3','Student4','Student5'])
print("DataFrame1=\n", dataFrame1)
dataFrame2 = pd.DataFrame(data2, index=["Student6", "Student7", "Student8"])
print("DataFrame2=\n", dataFrame1)
    
#CONCATENATE TWO DATAFRAME
resDF = pd.concat([dataFrame1,dataFrame2])
print("\nConcatenating DataFrames = \n", resDF)
```

```python
#APPEND NEW CATEGORIES IN PANDAS

import pandas as pd

#Create a categorical Series

s = pd.Series(data =  ["p", "q", "r", "s", "t"], dtype = "category")

#Display the series

print("Series = \n", s)

#Append a category

s = s.cat.add_categories("a")

#Display the updated category

print("\nUpdated category \n", s)
```

```python
#Remove a category in pandas:

import pandas as pd

#Create a categorical Series

s = pd.Series(data =  ["p", "q", "r", "s", "t"], dtype = "category")

#Display the series

print("Series = \n", s)

#Append a category

s = s.cat.remove_categories("q")

#Display the updated category

print("\nUpdated category \n", s)
```

```python
# Aggregation operation- Get the mean of the grouped data
```

```python
import pandas as pd
import numpy as np

#Data
data = {
  'Player': ["Amit", "John", "Amit", "David", "Steve", "John"],
  "Rank": [1, 4, 3, 5, 2, 7],
  'Year': [2023, 2022, 2021, 2022, 2023, 2019],
  'points': [95, 70, 65, 80, 90, 50]
}

# Our dataframe
df = pd.DataFrame(data)

#Display the records
print("Cricket Player Records\n", df)

# use the groupby() to group
groupRes = df.groupby('Year')

# Use the agg() to perform aggregation

print("Mean:\n", groupRes['points'].agg(np.mean))
```

```python
# Add a legend in a graph with Matplotlib

import matplotlib.pyplot as plt
import numpy as np

#Data
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9]

#Create plots
fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b, 'k--', label = 'Frequency')
ax.plot(a,c, 'k:', label = 'Periods')

# Create a lagend using the matplotlib Axes.legend() in python
ax.legend()

#plot Title
plt.title("Frequency of a signal")

#Display the result
plt.show()

```

```python
# Change the background color of the legend in matplotlib

import matplotlib.pyplot as plt
import numpy as np

#Data
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9]

#Create plots
fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b, 'k--', label = 'Frequency')
ax.plot(a,c, 'k:', label = 'Periods')

#Create the legend
legend = ax.legend(loc = "upper center")

#set the background
legend.get_frame().set_facecolor('violet')

#plot Title
plt.title("Frequency of a signal")

#Display the result
plt.show()
```

```python
# TO INCREASE OR DECREASE THE FONT SIZE OF LEGEND IN A GRAPH, USE THE FONTSIZE PARAMETER OF THE LEGEND() METHOD. THE VALUE OF THE frozenset CAN BE:
# XX=SMALL, X-SMALL, SMALL, MEDIUM, LARGE, X-LARGE, XX-LARGE
```

```python
# Change the font size of the legend in matplotlib

import matplotlib.pyplot as plt
import numpy as np

#Data
a = np.arange(5)
b = [2, 4, 6, 8, 10]
c = [5, 6, 7, 8, 9]

#Create plots
fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b, 'k--', label = 'Frequency')
ax.plot(a,c, 'k:', label = 'Periods')

#Create the legend
legend = ax.legend(loc = "upper center", fontsize = 'medium')

#set the background
legend.get_frame().set_facecolor('violet')

#plot Title
plt.title("Frequency of a signal")

#Display the result
plt.show()
```

```python
# 1- the 1st parameter is x, the wedge sizes are set here
# 2- the labels parmeter is the sequence od strings providing the lables for each wedge
# 3- the third parameter is autopct, a string that labels the wedges with their numeric value. tHE LABEL WILL BE PLACED INSIDE THE WEDGE. IF AUTOPCT IS A FORMAT STRING THE LABEL WILL BE FMT % PCT
```

```python
# PLOT A PIE CHART IN MATPLOTLIB
import matplotlib.pyplot as plt
import numpy as np

#Data
cricketer = np.array(["Tim","john","Amit","jocob","Karl","Gray",])
runs = np.array([85, 59, 99, 77, 67, 95])

# Plot a pie chart using the pie() method
plt.pie(runs, labels = cricketer, autopct = '%1.2f%%')

#Plot Title
plt.title("Runs scored in a cricket match")


#Display the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data to plot a Line

months = np.array(["Jan","Feb","Mar","Apr","May","June","July","Aug","Sept","Oct","Nov","Dec"])
rain = np.array([60, 55, 40, 45, 70, 80, 85, 70, 62, 79, 36, 67])

# Plot a line graph using the plot() method
# The two parameter are x and y axis

plt.plot(months, rain)

#Labels
plt.xlabel("Months")
plt.ylabel("Rainfall (mm)")

#Plot Title
plt.title("Rainfall in an Year")

#Display the figure
plt.show()
```
